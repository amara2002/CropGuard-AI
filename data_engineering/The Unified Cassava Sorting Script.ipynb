```python
import pandas as pd
import os
import shutil
import json

# 1. YOUR SPECIFIC PATHS
json_path = r"C:\Users\Amara Nyei\Desktop\Model_Training\cassava_data\label_num_to_disease_map.json"
csv_path = r"C:\Users\Amara Nyei\Desktop\Model_Training\cassava_data\train.csv"
images_source_dir = r"C:\Users\Amara Nyei\Desktop\Model_Training\cassava_data\train_images"

# 2. WHERE YOU WANT THE DATA TO GO (The folder for training)
master_crop_data = r"C:\Users\Amara Nyei\Desktop\Model_Training\crop_data"

# Load the labels
with open(json_path, 'r') as f:
    cassava_map = json.load(f)

# Load the CSV mapping
df = pd.read_csv(csv_path)

print(f"Starting to sort {len(df)} images...")

# Loop through and copy
for index, row in df.iterrows():
    # We clean the name to remove spaces and add a prefix
    disease_name = "Cassava_" + cassava_map[str(row['label'])].replace(" ", "_")
    
    # Define source and destination
    source_file = os.path.join(images_source_dir, row['image_id'])
    dest_folder = os.path.join(master_crop_data, disease_name)
    
    # Create the folder if it's not there
    os.makedirs(dest_folder, exist_ok=True)
    
    # Copy the image
    if os.path.exists(source_file):
        shutil.copy(source_file, os.path.join(dest_folder, row['image_id']))
    
    # Optional: Print progress every 1000 images
    if index % 1000 == 0:
        print(f"Processed {index} images...")

print(f"\nSuccess! Check your '{master_crop_data}' folder.")
print(f"Classes found: {list(cassava_map.values())}")
```
