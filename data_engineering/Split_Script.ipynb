```python
import os
import shutil
import random

# 1. SOURCE (This folder has EVERYTHING: Bean, Cassava, Corn, Potato, Tomato)
source_dir = r"C:\Users\Amara Nyei\Desktop\Model_Training\crop_data"

# 2. DESTINATION (Where the AI will actually look)
train_dir = r"C:\Users\Amara Nyei\Desktop\Model_Training\final_dataset\train"
val_dir = r"C:\Users\Amara Nyei\Desktop\Model_Training\final_dataset\val"

# Split 80% for Training, 20% for Validation
split_ratio = 0.8

# Get every single folder inside crop_data
all_folders = [f for f in os.listdir(source_dir) if os.path.isdir(os.path.join(source_dir, f))]

print(f"Found {len(all_folders)} different crop/disease categories. Starting split...")

for folder in all_folders:
    # Create the matching folders in 'train' and 'val'
    os.makedirs(os.path.join(train_dir, folder), exist_ok=True)
    os.makedirs(os.path.join(val_dir, folder), exist_ok=True)
    
    # Get all images in this specific folder
    all_images = os.listdir(os.path.join(source_dir, folder))
    random.shuffle(all_images)
    
    # Split the list
    limit = int(len(all_images) * split_ratio)
    train_list = all_images[:limit]
    val_list = all_images[limit:]
    
    # Copy images to Train
    for img in train_list:
        shutil.copy(os.path.join(source_dir, folder, img), os.path.join(train_dir, folder, img))
        
    # Copy images to Val
    for img in val_list:
        shutil.copy(os.path.join(source_dir, folder, img), os.path.join(val_dir, folder, img))
    
    print(f"✅ Processed: {folder}")

print("\nDONE! Your 'final_dataset' now contains ALL your crops together.")
```
